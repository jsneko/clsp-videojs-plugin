```python
%matplotlib inline

import pandas as pd
import os

path = '/some/path/to/file'

for filename in os.listdir(os.getcwd()):
    if filename[-4:] == '.csv':
        df = pd.read_csv(filename,
                         parse_dates=['(PDH-CSV 4.0) (Eastern Daylight Time)(240)'],
                         index_col="(PDH-CSV 4.0) (Eastern Daylight Time)(240)",
                         na_values=" ")
        depth = 10 * df.size
        if int(depth) > 1000:
            depth = 1000
        df.plot(figsize=(30, depth), subplots=True)
```
