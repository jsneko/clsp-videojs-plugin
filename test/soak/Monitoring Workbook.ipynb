```python
%matplotlib inline

import pandas as pd
import os
import math
import datetime
import dateutil
import numpy as np
import matplotlib.pyplot as plt

for filename in os.listdir(os.getcwd()):
    if filename[-4:] == '.csv':
        data_frame = pd.read_csv(filename,
                         parse_dates=['(PDH-CSV 4.0) (Eastern Daylight Time)(240)'],
                         index_col="(PDH-CSV 4.0) (Eastern Daylight Time)(240)",
                         na_values=" ")

        num_rows = math.ceil(len(data_frame.columns) / 3)

        fig, ax_array = plt.subplots(num_rows, 3, sharex=False, sharey=False)

        flat_ax_array = np.ravel(ax_array)
        
        for loop_index, data in enumerate(data_frame.iteritems()):

            column_name = data[0]
            x = data[1].index.time
            data[1].values.sort()
            y = data[1].values
            x_val = [val.microsecond for val in x]
            x_ticks = [val.minute for val in x]

            subAxes = flat_ax_array[loop_index]
            subAxes.plot(x_val, y)
            subAxes.set_xticklabels(x_ticks)
            
        fig.set_figheight(num_rows * 5)
        fig.set_figwidth(15)

```
